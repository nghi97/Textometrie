```python
import os
os.chdir("C://Users//User//Desktop//")
chemin='Textometrie.txt'
```

```python
import re            # module "regular expression" 
f=open(chemin,"r",encoding="utf-8")   
textebrut = f.read() 
textebrut
textchaine = re.sub(r"[.:' ;,?=!\n]+()"," ",textebrut) # remplacement de la ponctuation par des blancs, dont les retours à la ligne (\n)
textchaine
```

```python
#Separateur des articles
textes=textchaine.split("////")
textes
```

```python
#Separateur des categories et des articles
cate=[]
arti=[]
for i in range(len(textes)):
    cate.append(textes[i].split('****')[0])
    arti.append(textes[i].split('****')[1])
cate
```

```python
titre=[]
contenu=[]
for i in range(len(arti)):
    titre.append(arti[i].split('####')[0])
    contenu.append(arti[i].split('####')[1])
    
titre
contenu
```

```python
nt=len(titre)
nt
```

```python
d=' '.join(contenu)
d=d.lower()
```

```python
e=d.split()
e
```

```python
motot={}    # motot est un objet de type dictionnaire "mot + effectif"
for mot in e:
    motot[mot] = 0  
for  mot in e:
    motot[mot] += 1 
```

```python
motot
```

```python
motex = {}
for mot in motot:
    motex[mot] = [0]*nt   # chaque mot du dictionnaire est associé à une liste d'effectifs plutôt qu'à un effectif simple (en vue de la construction de la table lexicale)

for i in range(nt):   
    for mot in contenu[i].lower().split():
        motex[mot][i] += 1   # chaque mot est associé à son effectif dans chaque texte
```

```python
motex
```

```python
motex4={}
g=[]
for mot in motex:
    k=motex[mot]
    g.append([k[0]+k[1],k[2]+k[3],k[4]+k[5],k[6]+k[7]])

g   
```

```python
import pandas as pd

        
```

```python
motex4=pd.DataFrame(data=g, index=motex.keys(),columns=['CULTURE','INNOVATION & TECHNOLOGIE ',' VIOLENCE & CONFLIT ', ' ENVIRONNEMENT '])
```

```python
motex4.to_csv('Textometrie.csv')
```
